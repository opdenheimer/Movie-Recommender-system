```python
# now here we will try to extract top 3 names of cast 
def convert3(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=3:
             l.append(i['name'])
             counter+=1   
        else:
            break
        
      
    return l
```

```python

import ast

bad_rows = []

def fetch_director(obj, index=None):
    l = []
    try:
        for i in ast.literal_eval(obj):
            if i.get('job') == 'Director':
                l.append(i.get('name'))
        return l  
    except Exception as e:
        if index is not None:
            bad_rows.append(index) 
        return []  

```

```python
movies['test'] = movies['crew'].apply(lambda x: fetch_director(x))  # skip index
movies.sample(5)
```

```python
movies['crew'] = movies.apply(lambda row: fetch_director(row['crew'], row.name), axis=1)
```

```python
import nltk
```

```python
 #stemming
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
ps.stem('loved')
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y) 
```

```python
new_df['tags']= new_df['tags'].apply(stem)
```

```python
#vectorization 
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors= cv.fit_transform(new_df['tags']).toarray()
```

```python
type(vectors)
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity= cosine_similarity(vectors)

     
```

```python
similarity[2]
```

```python
#here we took out cos sim of a movie with other movie sorted it by keep its idx intact
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
#fetching indx of a movie 
def recommend(movie):
    movie_idx= new_df[new_df['title'] == movie].index[0]
    distances=similarity[movie_idx]
    movies_list= sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
        
       
    return    
```

```python
recommend('Avatar')
    
```
